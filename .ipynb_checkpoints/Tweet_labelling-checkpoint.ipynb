```python
# Python program to create a simple GUI
# Simple Quiz using Tkinter
 
#import everything from tkinter
from tkinter import *
 
# and import messagebox as mb from tkinter
from tkinter import messagebox as mb
 
#import json to use json file for data
import json
import pandas as pd
import os
 
#class to define the components of the GUI
class TweetLabel:
    # This is the first method which is called when a
    # new object of the class is initialized. This method
    # sets the question count to 0. and initialize all the
    # other methoods to display the content and make all the
    # functionalities available
    def __init__(self):
        
         
        # set tweet number to 0
        with open("log.txt","r") as log:
            l=log.read()
        try: 
            l=int(l)
            self.t_no=int(l)
        except:
            self.t_no=0
        self.data_size=len(question)
        
        self.label_csv = pd.read_csv(f'tweet_label_{User}.csv')
        self.label=["postive","neutral","negative"]
        
        # assigns ques to the display_question function to update later.
        self.display_title()
        self.t_label = Label(gui,width=600,font=( 'ariel' ,12, 'bold' ), anchor= 'w',wraplength=600)
        self.display_tweets()
         
        # opt_selected holds an integer value which is used for
        # selected option in a twitter.
        self.opt_selected=IntVar()
         
        # displaying radio button for the current question and used to
        # display options for the current question
        self.opts=self.radio_buttons()
         
        # display options for the current question
        self.display_options()
         
        # displays the button for next and exit.
        self.buttons()
        gui.bind("<Right>",self.next_btn)
        gui.bind("<Left>",self.prev_btn)
         
        # no of questions
        
    # This method checks the Answer after we click on Next.
    def write_label(self,t_no):        
        # checks 
        try:
            if self.opt_selected.get()!=0:
                if self.label[self.opt_selected.get()-1] == self.label[0]:

                    self.label_csv.loc[self.t_no,'tweets']=question[self.t_no]
                    self.label_csv.loc[self.t_no,'labels']= self.label[0]


                elif self.label[self.opt_selected.get()-1] == self.label[1]:
                    self.label_csv.loc[self.t_no, 'tweets']=question[self.t_no]
                    self.label_csv.loc[self.t_no,'labels']= self.label[1]

                elif self.label[self.opt_selected.get()-1] == self.label[2]:
                    self.label_csv.loc[self.t_no, 'tweets']=question[self.t_no]
                    self.label_csv.loc[self.t_no,'labels']= self.label[2]
            else:
                self.label_csv.loc[self.t_no, 'tweets']=question[self.t_no]
                self.label_csv.loc[self.t_no,'labels']= None
                

            
            # print(self.label_csv)
            
        except:
            print("ERROR, WHILE SELECTING OPTIONS")
    def save_label(self):
        self.label_csv.to_csv(f"tweet_label_{User}.csv")
           
    def next_btn(self,event=None):
        #Write the label into the file
        self.write_label(self.t_no)
        
        #Save Label
        self.save_label()
        
        # Moves to next twitter by incrementing the t_no counter
        self.t_no += 1
        
        # checks if the q_no size is equal to the data size
        if self.t_no==self.data_size:
            with open("log.txt","w") as log:
                log.write(str(self.label_csv.index[-1]))
            gui.destroy()
        else:
            # shows the next question
            self.display_tweets()
            self.display_options()
            
    def delete_label(self):
        # print(self.label_csv)
        self.label_csv.drop(index=self.label_csv.index[-1],axis=0,inplace=True)
        
    def prev_btn(self,event=None):
        #Write the label into the file
        try:
            self.delete_label()
        except:
            self.t_no=0  
        
        #Save Label
        self.save_label()
        
        # Moves to next twitter by incrementing the t_no counter
        if self.t_no>0:
            self.t_no -= 1
              
        # shows the next question
        self.display_tweets()
        self.display_options()
    def Quit(self):
        with open("log.txt","w") as log:
            log.write(str(self.label_csv.index[-1]))
        gui.destroy()
 
 
    
    def buttons(self):
         
        # The first button is the Next button to move to the
        # next twitter
        next_button = Button(gui, text="Next",command=self.next_btn,
        width=10,bg="blue",fg="white",font=("ariel",16,"bold"))
         
        # placing the button  on the screen
        next_button.place(x=600,y=380)
         
        # This is the second button which is used to Quit the GUI
        quit_button = Button(gui, text="Quit", command=self.Quit,
        width=5,bg="black", fg="white",font=("ariel",16," bold"))
        
        prev_button = Button(gui, text="prev",command=self.prev_btn,
        width=10,bg="blue",fg="white",font=("ariel",16,"bold"))
         
        # placing the button  on the screen
        prev_button.place(x=50,y=380)
         
        # placing the Quit button on the screen
        quit_button.place(x=700,y=50)
 
 
    def display_options(self):
        val=0
         
        # deselecting the options
        self.opt_selected.set(0)
         
        # looping over the options to be displayed for the
        # text of the radio buttons.
        for option in self.label:
            self.opts[val]['text']=option
            val+=1
 
 
    # This method shows the current Question on the screen
    def display_tweets(self):       
        # setting the Question properties
            
        # t_label = Label(gui, text=question[self.t_no], width=600,
        # font=( 'ariel' ,12, 'bold' ), anchor= 'w',wraplength=600 )
        self.t_label["text"]=question[self.t_no]
        #placing the option on the screen
        self.t_label.place(x=70, y=100)
    
 
 
    # This method is used to Display Title
    def display_title(self):
         
        # The title to be shown
        title = Label(gui, text="Tweets Labelling APP",width=50, bg="green",fg="white", font=("ariel", 20, "bold"))
         
        # place of the title
        title.place(x=0, y=2)
 
 
    # This method shows the radio buttons to select the Question
    # on the screen at the specified position. It also returns a
    # list of radio button which are later used to add the options to
    # them.
    def radio_buttons(self):
         
        # initialize the list with an empty list of options
        q_list = []
         
        # position of the first option
        y_pos = 200
         
        # adding the options to the list
        while len(q_list) < 3:
             
            # setting the radio button properties
            radio_btn = Radiobutton(gui,text=" ",variable=self.opt_selected,
            value = len(q_list)+1,font = ("ariel",14))
             
            # adding the button to the list
            q_list.append(radio_btn)
             
            # placing the button
            radio_btn.place(x = 100, y = y_pos)
             
            # incrementing the y-axis position by 40
            y_pos += 40
         
        # return the radio buttons
        return q_list
 

   
# Create a GUI Window
gui = Tk()

#EDIT THIS ACCORDING TO USER
User="Prashamsa"
PATH_TO_CSV="Final_tweets_msft.csv"

# set the size of the GUI Window
gui.geometry("800x450")
 
# set the title of the Window
gui.title("Twitter label")
 
# get the data from the json fileself.label_csv.loc[self.t_no,'labels']= self.label[0]
df = pd.read_csv(PATH_TO_CSV) 


if not os.path.exists("log.txt"):
    open("log.txt",'w').close()
    
if not os.path.exists(f"tweet_label_{User}.csv"):
    csv=pd.DataFrame(columns = ['tweets', 'labels'])
    csv.to_csv(f"tweet_label_{User}.csv")
    
# set the question, options, and answer
question =list(df["Tweets"])[0:2523]
# create an object of the Quiz Class.

Tweet = TweetLabel()


 
# Start the GUI
gui.mainloop()
 
# END OF THE PROGRAM
```
